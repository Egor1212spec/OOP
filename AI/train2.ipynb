```python
# -*- coding: utf-8 -*-
"""
Финальный скрипт сравнения.
1. Загружает уже обученные RuBERT-Base и RuBERT-Tiny (без повторного обучения).
2. Обучает с нуля УЛУЧШЕННЫЙ Custom Transformer (BERT-sized).
3. Сравнивает их на одном тестовом наборе.
"""

import os
import glob
import csv
import gc
import time
import warnings
from collections import Counter

import torch
import torch.nn as nn
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, f1_score

from datasets import Dataset
from transformers import (
    AutoTokenizer,
    AutoModelForSequenceClassification,
    Trainer,
    TrainingArguments,
    DataCollatorWithPadding,
    EarlyStoppingCallback
)
from torch.nn import CrossEntropyLoss

warnings.filterwarnings("ignore")

# ==========================================
# 1. НАСТРОЙКИ
# ==========================================
INPUT_DIR = 'Data'
OUTPUT_ROOT = './comparison_results'
SEED = 42
MAX_LENGTH = 256
DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# --- КОНФИГУРАЦИЯ СРАВНЕНИЯ ---
# Укажите здесь пути к уже обученным моделям
PRETRAINED_MODELS = [
    {
        "name": "RuBERT-Base (Fine-tuned)",
        "path": "./fine_tuned_rubert_base", # Путь к папке с сохраненной моделью
        "base_tokenizer": "DeepPavlov/rubert-base-cased"
    },
    {
        "name": "RuBERT-Tiny (Fine-tuned)",
        "path": "./fine_tuned_rubert_tiny", # Путь к папке с сохраненной моделью
        "base_tokenizer": "cointegrated/rubert-tiny"
    }
]

# Настройки для обучаемого с нуля Кастомного Трансформера
CUSTOM_MODEL_CONFIG = {
    "name": "Custom-BERT-Improved",
    "epochs": 15,              # Больше эпох для обучения с нуля
    "batch_size": 16,          # Батч
    "lr": 5e-5,                # Чуть выше LR для старта
    "embed_dim": 768,          # Размер как у BERT Base
    "heads": 12,
    "layers": 12,
    "ff_dim": 3072,
    "dropout": 0.1
}

os.makedirs(OUTPUT_ROOT, exist_ok=True)
print(f"Device: {DEVICE}")

# ==========================================
# 2. ПОДГОТОВКА ДАННЫХ
# ==========================================
def load_data():
    files = glob.glob(os.path.join(INPUT_DIR, 'output_*.txt'))
    if not files:
        raise FileNotFoundError(f"Файлы output_*.txt не найдены в {INPUT_DIR}")

    data = []
    unique_labels = set()
    
    # 1. Собираем уникальные метки
    for fp in files:
        try:
            lbl = int(os.path.basename(fp).split('_')[1].split('.')[0])
            unique_labels.add(lbl)
        except: continue
        
    sorted_labels = sorted(list(unique_labels))
    label_map = {orig: i for i, orig in enumerate(sorted_labels)}
    
    # 2. Читаем данные
    for fp in files:
        try:
            orig = int(os.path.basename(fp).split('_')[1].split('.')[0])
            if orig not in label_map: continue
            mapped = label_map[orig]
            with open(fp, 'r', encoding='utf-8') as f:
                reader = csv.reader(f)
                texts = next(reader)
                for t in texts:
                    if t.strip():
                        data.append({"text": t.strip(), "label": mapped})
        except: continue
        
    return data, label_map, sorted_labels

raw_data, label_map, sorted_labels = load_data()
num_classes = len(label_map)
print(f"Загружено {len(raw_data)} примеров, {num_classes} классов.")

# Split
labels_list = [d['label'] for d in raw_data]
train_raw, test_raw = train_test_split(
    raw_data, test_size=0.2, random_state=SEED, stratify=labels_list
)

# Веса классов
train_lbls = [d['label'] for d in train_raw]
cnt = Counter(train_lbls)
tot = len(train_lbls)
weights = [tot / (num_classes * cnt[i]) if cnt[i] > 0 else 0 for i in range(num_classes)]
class_weights_tensor = torch.tensor(weights, dtype=torch.float).to(DEVICE)
print("Class Weights:", [round(w, 2) for w in weights])

# ==========================================
# 3. КЛАССЫ УЛУЧШЕННОЙ МОДЕЛИ
# ==========================================

class BetterTransformerLayer(nn.Module):
    """ Слой с Pre-Layer Normalization для стабильности """
    def __init__(self, dim, heads, ff_dim, dropout=0.1):
        super().__init__()
        self.norm1 = nn.LayerNorm(dim)
        self.norm2 = nn.LayerNorm(dim)
        self.self_attn = nn.MultiheadAttention(embed_dim=dim, num_heads=heads, dropout=dropout, batch_first=True)
        self.ff = nn.Sequential(
            nn.Linear(dim, ff_dim),
            nn.GELU(),
            nn.Dropout(dropout),
            nn.Linear(ff_dim, dim)
        )
        self.dropout = nn.Dropout(dropout)

    def forward(self, x, key_padding_mask=None):
        x_norm = self.norm1(x)
        attn_out, _ = self.self_attn(x_norm, x_norm, x_norm, key_padding_mask=key_padding_mask)
        x = x + self.dropout(attn_out)
        x_norm = self.norm2(x)
        ff_out = self.ff(x_norm)
        x = x + self.dropout(ff_out)
        return x

class CustomTransformerModel(nn.Module):
    """ Кастомный трансформер с Mean Pooling и Xavier Init """
    def __init__(self, config, vocab_size, num_classes):
        super().__init__()
        self.config = config
        self.num_labels = num_classes
        embed_dim = config.get("embed_dim", 768)
        
        self.embed = nn.Embedding(vocab_size, embed_dim, padding_idx=0)
        self.pos_emb = nn.Parameter(torch.zeros(1, MAX_LENGTH, embed_dim))
        self.emb_norm = nn.LayerNorm(embed_dim)
        self.emb_dropout = nn.Dropout(config.get("dropout", 0.1))
        
        self.layers = nn.ModuleList([
            BetterTransformerLayer(embed_dim, config.get("heads", 12), config.get("ff_dim", 3072), config.get("dropout", 0.1))
            for _ in range(config.get("layers", 12))
        ])
        
        self.final_norm = nn.LayerNorm(embed_dim)
        self.classifier = nn.Linear(embed_dim, num_classes)
        
        self.apply(self._init_weights)

    def _init_weights(self, module):
        if isinstance(module, nn.Linear):
            nn.init.xavier_uniform_(module.weight)
            if module.bias is not None: nn.init.zeros_(module.bias)
        elif isinstance(module, nn.Embedding):
            nn.init.normal_(module.weight, mean=0.0, std=0.02)
        elif isinstance(module, nn.LayerNorm):
            nn.init.ones_(module.weight)
            nn.init.zeros_(module.bias)

    def forward(self, input_ids=None, attention_mask=None, labels=None, **kwargs):
        x = self.embed(input_ids) + self.pos_emb[:, :input_ids.size(1), :]
        x = self.emb_dropout(self.emb_norm(x))
        
        key_padding_mask = (input_ids == 0)
        for layer in self.layers:
            x = layer(x, key_padding_mask=key_padding_mask)
        x = self.final_norm(x)
        
        # Mean Pooling
        if attention_mask is not None:
            mask_expanded = attention_mask.unsqueeze(-1).expand(x.size()).float()
            sum_embeddings = torch.sum(x * mask_expanded, 1)
            sum_mask = torch.clamp(mask_expanded.sum(1), min=1e-9)
            pooled = sum_embeddings / sum_mask
        else:
            pooled = x.mean(dim=1)
            
        logits = self.classifier(pooled)
        
        loss = None
        if labels is not None:
            loss_fct = CrossEntropyLoss(weight=class_weights_tensor)
            loss = loss_fct(logits.view(-1, self.num_labels), labels.view(-1))
            
        return {"loss": loss, "logits": logits} if loss is not None else {"logits": logits}

# ==========================================
# 4. ФУНКЦИИ ОБУЧЕНИЯ И ОЦЕНКИ
# ==========================================

def compute_metrics(eval_pred):
    logits, labels = eval_pred
    preds = np.argmax(logits, axis=-1)
    return {
        "accuracy": accuracy_score(labels, preds),
        "f1_macro": f1_score(labels, preds, average="macro")
    }

class WeightedTrainer(Trainer):
    def compute_loss(self, model, inputs, return_outputs=False, **kwargs):
        labels = inputs.pop("labels")
        outputs = model(**inputs)
        logits = outputs.get("logits") if isinstance(outputs, dict) else outputs[1]
        loss_fct = CrossEntropyLoss(weight=class_weights_tensor)
        loss = loss_fct(logits.view(-1, num_classes), labels.view(-1))
        return (loss, outputs) if return_outputs else loss

def get_params_count(model):
    return sum(p.numel() for p in model.parameters())

results_summary = []

# --- 1. ОЦЕНКА ГОТОВЫХ МОДЕЛЕЙ (RUBERT) ---
print(f"\n{'='*20} ОЦЕНКА ГОТОВЫХ МОДЕЛЕЙ {'='*20}")

for model_info in PRETRAINED_MODELS:
    path = model_info["path"]
    name = model_info["name"]
    
    if not os.path.exists(path):
        print(f"[WARNING] Модель {name} не найдена по пути {path}. Пропускаем.")
        continue
        
    print(f"\nЗагрузка {name} из {path}...")
    try:
        tokenizer = AutoTokenizer.from_pretrained(path)
        model = AutoModelForSequenceClassification.from_pretrained(path)
        model.to(DEVICE)
    except Exception as e:
        print(f"Ошибка загрузки {name}: {e}. Пробуем загрузить базу {model_info['base_tokenizer']}...")
        # Fallback на базу (просто для теста кода, метрики будут плохие)
        tokenizer = AutoTokenizer.from_pretrained(model_info['base_tokenizer'])
        model = AutoModelForSequenceClassification.from_pretrained(model_info['base_tokenizer'], num_labels=num_classes)
        model.to(DEVICE)

    # Подготовка данных
    def tokenize_fn(batch):
        return tokenizer(batch["text"], truncation=True, max_length=MAX_LENGTH)
    
    test_ds = Dataset.from_list(test_raw).map(tokenize_fn, batched=True)
    test_ds.set_format("torch", columns=["input_ids", "attention_mask", "label"])
    test_ds = test_ds.rename_column("label", "labels")

    # Создаем Trainer только для evaluation
    trainer = Trainer(
        model=model,
        args=TrainingArguments(output_dir="./temp_eval", per_device_eval_batch_size=16, report_to="none"),
        compute_metrics=compute_metrics,
        eval_dataset=test_ds,
        data_collator=DataCollatorWithPadding(tokenizer)
    )
    
    start = time.time()
    metrics = trainer.evaluate()
    duration = time.time() - start
    
    params = get_params_count(model)
    print(f"--> Accuracy: {metrics['eval_accuracy']:.4f}, Params: {params:,}")
    
    results_summary.append({
        "Model": name,
        "Type": "Pre-trained",
        "Params": params,
        "Accuracy": metrics['eval_accuracy'],
        "F1 Macro": metrics['eval_f1_macro'],
        "Inference Time (s)": round(duration, 2)
    })
    
    del model, trainer, tokenizer
    torch.cuda.empty_cache()
    gc.collect()

# --- 2. ОБУЧЕНИЕ CUSTOM TRANSFORMER ---
print(f"\n{'='*20} ОБУЧЕНИЕ CUSTOM TRANSFORMER {'='*20}")

cfg = CUSTOM_MODEL_CONFIG
name = cfg["name"]

# Токенайзер берем от base, чтобы разбивал текст так же
tokenizer = AutoTokenizer.from_pretrained("DeepPavlov/rubert-base-cased")

def tokenize_fn(batch):
    return tokenizer(batch["text"], truncation=True, max_length=MAX_LENGTH)

train_ds = Dataset.from_list(train_raw).map(tokenize_fn, batched=True)
test_ds = Dataset.from_list(test_raw).map(tokenize_fn, batched=True)

cols = ["input_ids", "attention_mask", "label"]
train_ds.set_format("torch", columns=cols)
test_ds.set_format("torch", columns=cols)
train_ds = train_ds.rename_column("label", "labels")
test_ds = test_ds.rename_column("label", "labels")

# Инициализация модели
model = CustomTransformerModel(cfg, tokenizer.vocab_size, num_classes)
model.to(DEVICE)
params = get_params_count(model)
print(f"Архитектура: {cfg['layers']} layers, {cfg['embed_dim']} dim.")
print(f"Параметров: {params:,} (Сравнимо с BERT Base)")

# Аргументы обучения
training_args = TrainingArguments(
    output_dir=os.path.join(OUTPUT_ROOT, name),
    evaluation_strategy="epoch",
    save_strategy="epoch",
    learning_rate=cfg["lr"],
    per_device_train_batch_size=cfg["batch_size"],
    per_device_eval_batch_size=cfg["batch_size"],
    num_train_epochs=cfg["epochs"],
    weight_decay=0.01,
    warmup_ratio=0.1,        # Warmup критичен для обучения с нуля!
    load_best_model_at_end=True,
    metric_for_best_model="f1_macro",
    save_total_limit=1,
    fp16=torch.cuda.is_available(),
    logging_steps=50,
    report_to="none"
)

trainer = WeightedTrainer(
    model=model,
    args=training_args,
    train_dataset=train_ds,
    eval_dataset=test_ds,
    tokenizer=tokenizer,
    data_collator=DataCollatorWithPadding(tokenizer),
    compute_metrics=compute_metrics,
    callbacks=[EarlyStoppingCallback(early_stopping_patience=4)]
)

start = time.time()
trainer.train()
duration = time.time() - start

metrics = trainer.evaluate()

results_summary.append({
    "Model": name,
    "Type": "Trained from Scratch",
    "Params": params,
    "Accuracy": metrics['eval_accuracy'],
    "F1 Macro": metrics['eval_f1_macro'],
    "Inference Time (s)": round(duration, 2) # Тут это время обучения, пометим в таблице
})

# Сохраняем кастомную
trainer.save_model(os.path.join(OUTPUT_ROOT, name))

# ==========================================
# 5. ИТОГИ
# ==========================================
print(f"\n{'='*20} ИТОГОВАЯ ТАБЛИЦА {'='*20}")
df = pd.DataFrame(results_summary)
pd.options.display.float_format = '{:,.4f}'.format
print(df[["Model", "Params", "Accuracy", "F1 Macro"]])

# Сохранение CSV
df.to_csv(os.path.join(OUTPUT_ROOT, "final_comparison.csv"), index=False)

# График
plt.style.use('ggplot')
fig, ax = plt.subplots(figsize=(10, 6))
x = np.arange(len(df))
w = 0.35
ax.bar(x - w/2, df["F1 Macro"], w, label='F1 Macro', color='skyblue')
ax.bar(x + w/2, df["Accuracy"], w, label='Accuracy', color='orange')
ax.set_title('Сравнение: Pre-trained vs Custom (Improved)')
ax.set_xticks(x)
ax.set_xticklabels(df["Model"], rotation=15)
ax.legend()
plt.tight_layout()
plt.savefig(os.path.join(OUTPUT_ROOT, "comparison_chart.png"))

print(f"\nВсе готово! Результаты сохранены в {OUTPUT_ROOT}")
```
