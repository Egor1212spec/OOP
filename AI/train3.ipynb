```python
import glob
import csv
import gc
import time
import shutil
import random
import re
import warnings
import json
import math
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
warnings.filterwarnings("ignore")

import torch
import torch.nn as nn
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, f1_score
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.pipeline import Pipeline

from catboost import CatBoostClassifier, Pool
import pymorphy3
import nltk
nltk.download("stopwords", quiet=True)

from datasets import Dataset
from transformers import (
    AutoTokenizer,
    AutoModelForSequenceClassification,
    PreTrainedModel,
    PretrainedConfig,
    Trainer,
    TrainingArguments,
    DataCollatorWithPadding,
    EarlyStoppingCallback,
    TrainerCallback
)

INPUT_DIR = 'Data'
OUTPUT_ROOT = './models_tournament_final'
SEED = 42
MAX_LENGTH = 128
DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")

LOCAL_MODELS = {
    "ai-forever/ruBert-base": r"C:\Users\maths\.cache\huggingface\hub\models--ai-forever--ruBert-base\snapshots"
}

def get_local_model_path(base_path):
    if os.path.exists(base_path):
        snapshots = os.listdir(base_path)
        if snapshots:
            return os.path.join(base_path, snapshots[0])
    return None

AI_FOREVER_PATH = get_local_model_path(LOCAL_MODELS["ai-forever/ruBert-base"])

print("=" * 60)
print("–ö–û–ù–§–ò–ì–£–†–ê–¶–ò–Ø")
print("=" * 60)
print(f"PyTorch: {torch.__version__}")
print(f"Transformers: {__import__('transformers').__version__}")
print(f"–£—Å—Ç—Ä–æ–π—Å—Ç–≤–æ: {DEVICE}")
print(f"AI-Forever: {'‚úÖ ' + AI_FOREVER_PATH if AI_FOREVER_PATH else '‚ùå –ù–µ –Ω–∞–π–¥–µ–Ω–∞'}")
print("=" * 60)

morph = pymorphy3.MorphAnalyzer()

class CustomTransformerConfig(PretrainedConfig):
    model_type = "custom_transformer"
    
    def __init__(
        self,
        vocab_size=30000,
        hidden_size=256,
        num_hidden_layers=6,
        num_attention_heads=8,
        intermediate_size=1024,
        hidden_dropout_prob=0.1,
        attention_probs_dropout_prob=0.1,
        max_position_embeddings=512,
        num_labels=2,
        pad_token_id=0,
        **kwargs
    ):
        super().__init__(pad_token_id=pad_token_id, **kwargs)
        self.vocab_size = vocab_size
        self.hidden_size = hidden_size
        self.num_hidden_layers = num_hidden_layers
        self.num_attention_heads = num_attention_heads
        self.intermediate_size = intermediate_size
        self.hidden_dropout_prob = hidden_dropout_prob
        self.attention_probs_dropout_prob = attention_probs_dropout_prob
        self.max_position_embeddings = max_position_embeddings
        self.num_labels = num_labels

class MultiHeadAttention(nn.Module):
    def __init__(self, config):
        super().__init__()
        self.num_heads = config.num_attention_heads
        self.head_dim = config.hidden_size // config.num_attention_heads
        self.scale = self.head_dim ** -0.5
        
        self.q_proj = nn.Linear(config.hidden_size, config.hidden_size)
        self.k_proj = nn.Linear(config.hidden_size, config.hidden_size)
        self.v_proj = nn.Linear(config.hidden_size, config.hidden_size)
        self.out_proj = nn.Linear(config.hidden_size, config.hidden_size)
        self.dropout = nn.Dropout(config.attention_probs_dropout_prob)
    
    def forward(self, hidden_states, attention_mask=None):
        batch_size, seq_len, _ = hidden_states.size()
        
        q = self.q_proj(hidden_states).view(batch_size, seq_len, self.num_heads, self.head_dim).transpose(1, 2)
        k = self.k_proj(hidden_states).view(batch_size, seq_len, self.num_heads, self.head_dim).transpose(1, 2)
        v = self.v_proj(hidden_states).view(batch_size, seq_len, self.num_heads, self.head_dim).transpose(1, 2)
        
        attn_weights = torch.matmul(q, k.transpose(-2, -1)) * self.scale
        
        if attention_mask is not None:
            attn_weights = attn_weights + attention_mask
        
        attn_weights = torch.softmax(attn_weights, dim=-1)
        attn_weights = self.dropout(attn_weights)
        
        attn_output = torch.matmul(attn_weights, v)
        attn_output = attn_output.transpose(1, 2).contiguous().view(batch_size, seq_len, -1)
        
        return self.out_proj(attn_output)

class TransformerBlock(nn.Module):
    def __init__(self, config):
        super().__init__()
        self.attention = MultiHeadAttention(config)
        self.ln1 = nn.LayerNorm(config.hidden_size)
        self.ln2 = nn.LayerNorm(config.hidden_size)
        self.ffn = nn.Sequential(
            nn.Linear(config.hidden_size, config.intermediate_size),
            nn.GELU(),
            nn.Dropout(config.hidden_dropout_prob),
            nn.Linear(config.intermediate_size, config.hidden_size),
            nn.Dropout(config.hidden_dropout_prob)
        )
    
    def forward(self, hidden_states, attention_mask=None):
        attn_output = self.attention(self.ln1(hidden_states), attention_mask)
        hidden_states = hidden_states + attn_output
        ffn_output = self.ffn(self.ln2(hidden_states))
        hidden_states = hidden_states + ffn_output
        return hidden_states

class CustomTransformerForSequenceClassification(PreTrainedModel):
    config_class = CustomTransformerConfig
    
    def __init__(self, config):
        super().__init__(config)
        self.config = config
        
        self.embeddings = nn.Embedding(config.vocab_size, config.hidden_size, padding_idx=config.pad_token_id)
        self.position_embeddings = nn.Embedding(config.max_position_embeddings, config.hidden_size)
        self.embedding_dropout = nn.Dropout(config.hidden_dropout_prob)
        
        self.layers = nn.ModuleList([TransformerBlock(config) for _ in range(config.num_hidden_layers)])
        self.ln_final = nn.LayerNorm(config.hidden_size)
        
        self.classifier = nn.Sequential(
            nn.Linear(config.hidden_size, config.hidden_size),
            nn.Tanh(),
            nn.Dropout(config.hidden_dropout_prob),
            nn.Linear(config.hidden_size, config.num_labels)
        )
        
        self.init_weights()
    
    def forward(self, input_ids, attention_mask=None, labels=None, **kwargs):
        batch_size, seq_len = input_ids.size()
        
        position_ids = torch.arange(seq_len, device=input_ids.device).unsqueeze(0).expand(batch_size, -1)
        
        hidden_states = self.embeddings(input_ids) + self.position_embeddings(position_ids)
        hidden_states = self.embedding_dropout(hidden_states)
        
        if attention_mask is not None:
            extended_mask = (1.0 - attention_mask.unsqueeze(1).unsqueeze(2)) * -10000.0
        else:
            extended_mask = None
        
        for layer in self.layers:
            hidden_states = layer(hidden_states, extended_mask)
        
        hidden_states = self.ln_final(hidden_states)
        pooled = hidden_states[:, 0]
        logits = self.classifier(pooled)
        
        loss = None
        if labels is not None:
            loss_fn = nn.CrossEntropyLoss()
            loss = loss_fn(logits, labels)
        
        return {"loss": loss, "logits": logits} if loss is not None else {"logits": logits}

if os.path.exists(OUTPUT_ROOT):
    shutil.rmtree(OUTPUT_ROOT)
os.makedirs(OUTPUT_ROOT, exist_ok=True)

random.seed(SEED)
np.random.seed(SEED)
torch.manual_seed(SEED)
if torch.cuda.is_available():
    torch.cuda.manual_seed_all(SEED)

def preprocess_text(text):
    text_clean = re.sub(r'[^\w\s]', '', text).lower()
    return " ".join([morph.parse(w)[0].normal_form for w in text_clean.split()])

def load_data():
    files = glob.glob(os.path.join(INPUT_DIR, 'output_*.txt'))
    data = []
    labels_set = set()
    
    for fp in files:
        try:
            label = int(os.path.basename(fp).split('_')[1].split('.')[0])
            labels_set.add(label)
        except:
            continue
    
    sorted_labels = sorted(list(labels_set))
    label_map = {lbl: i for i, lbl in enumerate(sorted_labels)}
    
    print(f"[{time.strftime('%H:%M:%S')}] –ß—Ç–µ–Ω–∏–µ –∏ –ª–µ–º–º–∞—Ç–∏–∑–∞—Ü–∏—è –¥–∞–Ω–Ω—ã—Ö...")
    
    for fp in files:
        try:
            label = int(os.path.basename(fp).split('_')[1].split('.')[0])
            if label not in label_map:
                continue
            mapped_label = label_map[label]
            
            with open(fp, 'r', encoding='utf-8') as f:
                reader = csv.reader(f)
                texts = next(reader)
                for text in texts:
                    if text.strip():
                        lemma_text = preprocess_text(text.strip())
                        data.append({
                            "text_raw": text.strip(),
                            "text_lemma": lemma_text,
                            "label": mapped_label
                        })
        except:
            continue
    
    return data, label_map

raw_data, label_map = load_data()
num_classes = len(label_map)
print(f"[{time.strftime('%H:%M:%S')}] ‚úÖ –ó–∞–≥—Ä—É–∂–µ–Ω–æ {len(raw_data)} –ø—Ä–∏–º–µ—Ä–æ–≤, {num_classes} –∫–ª–∞—Å—Å–æ–≤")

indices = list(range(len(raw_data)))
labels_list = [d['label'] for d in raw_data]
train_idx, test_idx = train_test_split(
    indices, test_size=0.2, random_state=SEED, stratify=labels_list
)

def get_slice(indices, key):
    return [raw_data[i][key] for i in indices]

train_labels = get_slice(train_idx, 'label')
test_labels = get_slice(test_idx, 'label')

results = []
history_logs = {}

MODELS_CONFIG = [
    {
        "name": "LogisticRegression_TF-IDF",
        "type": "classical",
        "use_lemma": True,
        "model_class": LogisticRegression,
        "params": {"random_state": SEED, "max_iter": 1000}
    },
    {
        "name": "DecisionTree_TF-IDF",
        "type": "classical",
        "use_lemma": True,
        "model_class": DecisionTreeClassifier,
        "params": {"random_state": SEED}
    },
    {
        "name": "GradientBoosting_TF-IDF",
        "type": "classical",
        "use_lemma": True,
        "model_class": GradientBoostingClassifier,
        "params": {"random_state": SEED, "n_estimators": 100}
    },
    {
        "name": "CatBoost_TF-IDF",
        "type": "classical",
        "use_lemma": True,
        "model_class": CatBoostClassifier,
        "params": {
            "random_seed": SEED,
            "iterations": 500,
            "verbose": 0,
            "allow_writing_files": False,
            "task_type": "GPU" if torch.cuda.is_available() else "CPU"
        }
    },
    {
        "name": "CatBoost_Native_Text",
        "type": "catboost_native",
        "use_lemma": True,
        "params": {
            "random_seed": SEED,
            "iterations": 500,
            "verbose": 0,
            "allow_writing_files": False,
            "task_type": "GPU" if torch.cuda.is_available() else "CPU",
            "text_features": [0]
        }
    },
    {
        "name": "RuBERT_Tiny2_Lemma",
        "type": "transformer",
        "model_path": "cointegrated/rubert-tiny2",
        "use_lemma": True,
        "epochs": 10,
        "batch_size": 32
    },
    {
        "name": "Custom_Transformer_6L",
        "type": "custom_transformer",
        "use_lemma": True,
        "num_layers": 6,
        "hidden_size": 256,
        "num_heads": 8,
        "epochs": 15,
        "batch_size": 32
    },
    {
        "name": "Custom_Transformer_12L",
        "type": "custom_transformer",
        "use_lemma": True,
        "num_layers": 12,
        "hidden_size": 256,
        "num_heads": 8,
        "epochs": 15,
        "batch_size": 32
    },
]

if AI_FOREVER_PATH:
    MODELS_CONFIG.append({
        "name": "AI_Forever_RuBERT_Base_Lemma",
        "type": "transformer",
        "model_path": AI_FOREVER_PATH,
        "use_lemma": True,
        "epochs": 5,
        "batch_size": 4
    })

class PrinterCallback(TrainerCallback):
    def on_evaluate(self, args, state, control, metrics=None, **kwargs):
        if metrics:
            loss = metrics.get('eval_loss', 0)
            f1 = metrics.get('eval_f1_macro', 0)
            print(f"      Epoch {state.epoch:.0f}: Loss={loss:.4f}, F1={f1:.4f}")

def make_contiguous(model):
    for name, param in model.named_parameters():
        if not param.data.is_contiguous():
            param.data = param.data.contiguous()
    for name, buffer in model.named_buffers():
        if not buffer.is_contiguous():
            buffer.data = buffer.contiguous()
    return model

def build_vocab(texts, max_vocab=30000):
    word_counts = {}
    for text in texts:
        for word in text.split():
            word_counts[word] = word_counts.get(word, 0) + 1
    
    sorted_words = sorted(word_counts.items(), key=lambda x: x[1], reverse=True)
    vocab = {"[PAD]": 0, "[UNK]": 1, "[CLS]": 2, "[SEP]": 3}
    
    for word, _ in sorted_words[:max_vocab - len(vocab)]:
        vocab[word] = len(vocab)
    
    return vocab

def simple_tokenize(texts, vocab, max_length=128):
    input_ids_list = []
    attention_mask_list = []
    
    for text in texts:
        words = text.split()[:max_length - 2]
        tokens = [vocab.get("[CLS]", 2)]
        tokens += [vocab.get(w, vocab.get("[UNK]", 1)) for w in words]
        tokens.append(vocab.get("[SEP]", 3))
        
        attention_mask = [1] * len(tokens)
        padding_length = max_length - len(tokens)
        tokens += [vocab.get("[PAD]", 0)] * padding_length
        attention_mask += [0] * padding_length
        
        input_ids_list.append(tokens)
        attention_mask_list.append(attention_mask)
    
    return {
        "input_ids": torch.tensor(input_ids_list),
        "attention_mask": torch.tensor(attention_mask_list)
    }

def train_classical(config):
    print(f"\n[{time.strftime('%H:%M:%S')}] === {config['name']} ===")
    
    text_key = "text_lemma" if config.get("use_lemma", True) else "text_raw"
    X_train = get_slice(train_idx, text_key)
    X_test = get_slice(test_idx, text_key)
    
    start_time = time.time()
    
    if config['type'] == 'catboost_native':
        train_pool = Pool(data=[[t] for t in X_train], label=train_labels, text_features=[0])
        test_pool = Pool(data=[[t] for t in X_test], label=test_labels, text_features=[0])
        model = CatBoostClassifier(**config['params'])
        model.fit(train_pool, eval_set=test_pool, early_stopping_rounds=50)
        preds = model.predict(test_pool)
        n_params = model.tree_count_ * 1000
        
    elif config['model_class'].__name__ == 'CatBoostClassifier':
        vectorizer = TfidfVectorizer(max_features=5000)
        X_train_vec = vectorizer.fit_transform(X_train)
        X_test_vec = vectorizer.transform(X_test)
        
        model = CatBoostClassifier(**config['params'])
        model.fit(X_train_vec, train_labels, eval_set=(X_test_vec, test_labels),
                  early_stopping_rounds=50, verbose=False)
        preds = model.predict(X_test_vec)
        n_params = model.tree_count_ * 1000
        
    else:
        vectorizer = TfidfVectorizer(max_features=5000)
        model = config['model_class'](**config['params'])
        pipe = Pipeline([('vectorizer', vectorizer), ('classifier', model)])
        pipe.fit(X_train, train_labels)
        preds = pipe.predict(X_test)
        n_params = model.coef_.size if hasattr(model, 'coef_') else 0
    
    if hasattr(preds, 'shape') and len(preds.shape) > 1:
        preds = preds.flatten()
    preds = np.array(preds).astype(int)
    
    train_time = time.time() - start_time
    acc = accuracy_score(test_labels, preds)
    f1 = f1_score(test_labels, preds, average='macro')
    
    print(f"      –†–µ–∑—É–ª—å—Ç–∞—Ç: F1={f1:.4f}, Accuracy={acc:.4f}, Time={train_time:.1f}s")
    
    return {
        "Model": config['name'],
        "Type": config['type'],
        "Params": n_params,
        "Accuracy": round(acc, 4),
        "F1 Macro": round(f1, 4),
        "Time (s)": round(train_time, 1)
    }

def train_transformer(config):
    print(f"\n[{time.strftime('%H:%M:%S')}] === {config['name']} ===")
    
    model_path = config['model_path']
    model_dir = os.path.join(OUTPUT_ROOT, config['name'].replace(" ", "_"))
    use_lemma = config.get("use_lemma", True)
    text_key = "text_lemma" if use_lemma else "text_raw"
    
    print(f"      –ü—É—Ç—å: {model_path}")
    print(f"      –õ–µ–º–º–∞—Ç–∏–∑–∞—Ü–∏—è: {'–î–∞' if use_lemma else '–ù–µ—Ç'}")
    
    try:
        print(f"      –ó–∞–≥—Ä—É–∑–∫–∞ —Ç–æ–∫–µ–Ω–∏–∑–∞—Ç–æ—Ä–∞...")
        tokenizer = AutoTokenizer.from_pretrained(model_path)
        
        print(f"      –ó–∞–≥—Ä—É–∑–∫–∞ –º–æ–¥–µ–ª–∏...")
        model = AutoModelForSequenceClassification.from_pretrained(
            model_path,
            num_labels=num_classes,
            ignore_mismatched_sizes=True
        )
        model = make_contiguous(model)
        print(f"      ‚úÖ –ú–æ–¥–µ–ª—å –∑–∞–≥—Ä—É–∂–µ–Ω–∞")
        
    except Exception as e:
        print(f"      ‚ùå –û—à–∏–±–∫–∞: {e}")
        return None
    
    model.to(DEVICE)
    params_count = sum(p.numel() for p in model.parameters())
    print(f"      –ü–∞—Ä–∞–º–µ—Ç—Ä–æ–≤: {params_count:,}")
    
    train_texts = get_slice(train_idx, text_key)
    test_texts = get_slice(test_idx, text_key)
    
    def tokenize_fn(batch):
        return tokenizer(batch["text"], truncation=True, max_length=MAX_LENGTH)
    
    train_ds = Dataset.from_dict({"text": train_texts, "label": train_labels})
    test_ds = Dataset.from_dict({"text": test_texts, "label": test_labels})
    
    train_ds = train_ds.map(tokenize_fn, batched=True, desc="Tokenize train")
    test_ds = test_ds.map(tokenize_fn, batched=True, desc="Tokenize test")
    
    train_ds.set_format("torch", columns=['input_ids', 'attention_mask', 'label'])
    test_ds.set_format("torch", columns=['input_ids', 'attention_mask', 'label'])
    
    def compute_metrics(eval_pred):
        preds = np.argmax(eval_pred.predictions, axis=1)
        return {
            "accuracy": accuracy_score(eval_pred.label_ids, preds),
            "f1_macro": f1_score(eval_pred.label_ids, preds, average='macro')
        }
    
    training_args = TrainingArguments(
        output_dir=model_dir,
        learning_rate=5e-5,
        per_device_train_batch_size=config['batch_size'],
        per_device_eval_batch_size=config['batch_size'],
        num_train_epochs=config['epochs'],
        weight_decay=0.01,
        evaluation_strategy="epoch",
        save_strategy="epoch",
        save_total_limit=1,
        load_best_model_at_end=True,
        metric_for_best_model="f1_macro",
        greater_is_better=True,
        report_to="none",
        logging_steps=50,
        fp16=torch.cuda.is_available(),
        dataloader_num_workers=0,
        gradient_accumulation_steps=2 if config['batch_size'] <= 4 else 1,
        save_safetensors=False
    )
    
    trainer = Trainer(
        model=model,
        args=training_args,
        train_dataset=train_ds,
        eval_dataset=test_ds,
        tokenizer=tokenizer,
        data_collator=DataCollatorWithPadding(tokenizer),
        compute_metrics=compute_metrics,
        callbacks=[EarlyStoppingCallback(early_stopping_patience=3), PrinterCallback()]
    )
    
    start_time = time.time()
    trainer.train()
    train_time = time.time() - start_time
    
    eval_result = trainer.evaluate()
    history_logs[config['name']] = [log for log in trainer.state.log_history if 'eval_loss' in log]
    
    trainer.model = make_contiguous(trainer.model)
    trainer.save_model(model_dir)
    tokenizer.save_pretrained(model_dir)
    
    acc = eval_result['eval_accuracy']
    f1 = eval_result['eval_f1_macro']
    
    print(f"      –ò—Ç–æ–≥: F1={f1:.4f}, Accuracy={acc:.4f}, Time={train_time:.1f}s")
    
    del model, trainer
    torch.cuda.empty_cache()
    gc.collect()
    
    return {
        "Model": config['name'],
        "Type": "transformer",
        "Params": params_count,
        "Accuracy": round(acc, 4),
        "F1 Macro": round(f1, 4),
        "Time (s)": round(train_time, 1)
    }

def train_custom_transformer(config):
    print(f"\n[{time.strftime('%H:%M:%S')}] === {config['name']} ===")
    
    model_dir = os.path.join(OUTPUT_ROOT, config['name'].replace(" ", "_"))
    use_lemma = config.get("use_lemma", True)
    text_key = "text_lemma" if use_lemma else "text_raw"
    
    print(f"      –°–ª–æ–µ–≤: {config['num_layers']}, Hidden: {config['hidden_size']}, Heads: {config['num_heads']}")
    print(f"      –õ–µ–º–º–∞—Ç–∏–∑–∞—Ü–∏—è: {'–î–∞' if use_lemma else '–ù–µ—Ç'}")
    
    train_texts = get_slice(train_idx, text_key)
    test_texts = get_slice(test_idx, text_key)
    
    print(f"      –ü–æ—Å—Ç—Ä–æ–µ–Ω–∏–µ —Å–ª–æ–≤–∞—Ä—è...")
    vocab = build_vocab(train_texts, max_vocab=30000)
    print(f"      –†–∞–∑–º–µ—Ä —Å–ª–æ–≤–∞—Ä—è: {len(vocab)}")
    
    print(f"      –¢–æ–∫–µ–Ω–∏–∑–∞—Ü–∏—è...")
    train_tokens = simple_tokenize(train_texts, vocab, MAX_LENGTH)
    test_tokens = simple_tokenize(test_texts, vocab, MAX_LENGTH)
    
    train_ds = Dataset.from_dict({
        "input_ids": train_tokens["input_ids"].tolist(),
        "attention_mask": train_tokens["attention_mask"].tolist(),
        "label": train_labels
    })
    test_ds = Dataset.from_dict({
        "input_ids": test_tokens["input_ids"].tolist(),
        "attention_mask": test_tokens["attention_mask"].tolist(),
        "label": test_labels
    })
    
    train_ds.set_format("torch", columns=['input_ids', 'attention_mask', 'label'])
    test_ds.set_format("torch", columns=['input_ids', 'attention_mask', 'label'])
    
    model_config = CustomTransformerConfig(
        vocab_size=len(vocab),
        hidden_size=config['hidden_size'],
        num_hidden_layers=config['num_layers'],
        num_attention_heads=config['num_heads'],
        intermediate_size=config['hidden_size'] * 4,
        max_position_embeddings=MAX_LENGTH,
        num_labels=num_classes
    )
    
    model = CustomTransformerForSequenceClassification(model_config)
    model.to(DEVICE)
    
    params_count = sum(p.numel() for p in model.parameters())
    print(f"      –ü–∞—Ä–∞–º–µ—Ç—Ä–æ–≤: {params_count:,}")
    
    def compute_metrics(eval_pred):
        preds = np.argmax(eval_pred.predictions, axis=1)
        return {
            "accuracy": accuracy_score(eval_pred.label_ids, preds),
            "f1_macro": f1_score(eval_pred.label_ids, preds, average='macro')
        }
    
    def custom_collator(features):
        batch = {
            "input_ids": torch.stack([torch.tensor(f["input_ids"]) for f in features]),
            "attention_mask": torch.stack([torch.tensor(f["attention_mask"]) for f in features]),
            "labels": torch.tensor([f["label"] for f in features])
        }
        return batch
    
    training_args = TrainingArguments(
        output_dir=model_dir,
        learning_rate=1e-4,
        per_device_train_batch_size=config['batch_size'],
        per_device_eval_batch_size=config['batch_size'],
        num_train_epochs=config['epochs'],
        weight_decay=0.01,
        evaluation_strategy="epoch",
        save_strategy="epoch",
        save_total_limit=1,
        load_best_model_at_end=True,
        metric_for_best_model="f1_macro",
        greater_is_better=True,
        report_to="none",
        logging_steps=50,
        fp16=torch.cuda.is_available(),
        dataloader_num_workers=0,
        warmup_ratio=0.1,
        save_safetensors=False
    )
    
    trainer = Trainer(
        model=model,
        args=training_args,
        train_dataset=train_ds,
        eval_dataset=test_ds,
        data_collator=custom_collator,
        compute_metrics=compute_metrics,
        callbacks=[EarlyStoppingCallback(early_stopping_patience=5), PrinterCallback()]
    )
    
    start_time = time.time()
    trainer.train()
    train_time = time.time() - start_time
    
    eval_result = trainer.evaluate()
    history_logs[config['name']] = [log for log in trainer.state.log_history if 'eval_loss' in log]
    
    os.makedirs(model_dir, exist_ok=True)
    trainer.save_model(model_dir)
    with open(os.path.join(model_dir, "vocab.json"), "w", encoding='utf-8') as f:
        json.dump(vocab, f, ensure_ascii=False)
    
    acc = eval_result['eval_accuracy']
    f1 = eval_result['eval_f1_macro']
    
    print(f"      –ò—Ç–æ–≥: F1={f1:.4f}, Accuracy={acc:.4f}, Time={train_time:.1f}s")
    
    del model, trainer
    torch.cuda.empty_cache()
    gc.collect()
    
    return {
        "Model": config['name'],
        "Type": f"custom_transformer_{config['num_layers']}L",
        "Params": params_count,
        "Accuracy": round(acc, 4),
        "F1 Macro": round(f1, 4),
        "Time (s)": round(train_time, 1)
    }

print("\n" + "=" * 60)
print("üèÜ –¢–£–†–ù–ò–† –ú–û–î–ï–õ–ï–ô")
print("=" * 60)
print(f"Train: {len(train_idx)} | Test: {len(test_idx)} | –ö–ª–∞—Å—Å–æ–≤: {num_classes}")
print(f"–í—Å–µ –º–æ–¥–µ–ª–∏ –∏—Å–ø–æ–ª—å–∑—É—é—Ç –ª–µ–º–º–∞—Ç–∏–∑–∏—Ä–æ–≤–∞–Ω–Ω—ã–µ —Ç–µ–∫—Å—Ç—ã")
print("=" * 60)

for config in MODELS_CONFIG:
    try:
        if config['type'] in ['classical', 'catboost_native']:
            result = train_classical(config)
        elif config['type'] == 'transformer':
            result = train_transformer(config)
        elif config['type'] == 'custom_transformer':
            result = train_custom_transformer(config)
        else:
            continue
        
        if result:
            results.append(result)
            
    except Exception as e:
        print(f"\n‚ùå –û–®–ò–ë–ö–ê –≤ {config['name']}: {e}")
        import traceback
        traceback.print_exc()

if results:
    df_results = pd.DataFrame(results)
    df_results = df_results.sort_values(by="F1 Macro", ascending=False)
    
    print("\n" + "=" * 60)
    print("üìä –ò–¢–û–ì–û–í–ê–Ø –¢–ê–ë–õ–ò–¶–ê")
    print("=" * 60)
    print(df_results.to_string(index=False))
    
    df_results.to_csv(os.path.join(OUTPUT_ROOT, "results.csv"), index=False)
    
    if history_logs:
        plt.figure(figsize=(12, 6))
        for name, logs in history_logs.items():
            if logs:
                epochs = [log['epoch'] for log in logs]
                f1_scores = [log['eval_f1_macro'] for log in logs]
                plt.plot(epochs, f1_scores, marker='o', label=name, linewidth=2)
        
        plt.title("–ö—Ä–∏–≤—ã–µ –æ–±—É—á–µ–Ω–∏—è —Ç—Ä–∞–Ω—Å—Ñ–æ—Ä–º–µ—Ä–æ–≤ (F1 Macro)", fontsize=14)
        plt.xlabel("–≠–ø–æ—Ö–∞", fontsize=12)
        plt.ylabel("F1 Macro", fontsize=12)
        plt.legend(loc='lower right')
        plt.grid(True, alpha=0.3)
        plt.tight_layout()
        plt.savefig(os.path.join(OUTPUT_ROOT, "learning_curves.png"), dpi=150)
        print(f"\nüìà –ì—Ä–∞—Ñ–∏–∫: {OUTPUT_ROOT}/learning_curves.png")
    
    with open(os.path.join(OUTPUT_ROOT, "label_map.json"), "w", encoding='utf-8') as f:
        json.dump(label_map, f, ensure_ascii=False, indent=2)
    
    best = df_results.iloc[0]
    print(f"\nü•á –õ–£–ß–®–ê–Ø –ú–û–î–ï–õ–¨: {best['Model']}")
    print(f"   F1 Macro: {best['F1 Macro']}")
    print(f"   Accuracy: {best['Accuracy']}")

else:
    print("\n‚ùå –ù–µ—Ç —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–≤")

print("\n" + "=" * 60)
print("‚úÖ –¢–£–†–ù–ò–† –ó–ê–í–ï–†–®–Å–ù")
print("=" * 60)
```

```python

import os
import glob
import csv
import time
import re
import json
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from catboost import CatBoostClassifier, Pool
import pymorphy3
from sklearn.metrics import accuracy_score, f1_score, classification_report, confusion_matrix

INPUT_DIR = 'Data'
MODEL_DIR = './best_model_catboost'
SEED = 42

os.makedirs(MODEL_DIR, exist_ok=True)

morph = pymorphy3.MorphAnalyzer()

def preprocess_text(text):
    text_clean = re.sub(r'[^\w\s]', '', text).lower()
    return " ".join([morph.parse(w)[0].normal_form for w in text_clean.split()])

def load_all_data():
    files = glob.glob(os.path.join(INPUT_DIR, 'output_*.txt'))
    data = []
    labels_set = set()
    
    for fp in files:
        try:
            label = int(os.path.basename(fp).split('_')[1].split('.')[0])
            labels_set.add(label)
        except:
            continue
    
    sorted_labels = sorted(list(labels_set))
    label_map = {lbl: i for i, lbl in enumerate(sorted_labels)}
    reverse_label_map = {i: lbl for lbl, i in label_map.items()}
    
    print(f"[{time.strftime('%H:%M:%S')}] –ó–∞–≥—Ä—É–∑–∫–∞ –∏ –ª–µ–º–º–∞—Ç–∏–∑–∞—Ü–∏—è –≤—Å–µ—Ö –¥–∞–Ω–Ω—ã—Ö...")
    
    for fp in files:
        try:
            label = int(os.path.basename(fp).split('_')[1].split('.')[0])
            if label not in label_map:
                continue
            mapped_label = label_map[label]
            
            with open(fp, 'r', encoding='utf-8') as f:
                reader = csv.reader(f)
                texts = next(reader)
                for text in texts:
                    if text.strip():
                        data.append({
                            "text_raw": text.strip(),
                            "text_lemma": preprocess_text(text.strip()),
                            "label": mapped_label,
                            "original_label": label
                        })
        except:
            continue
    
    return data, label_map, reverse_label_map

data, label_map, reverse_label_map = load_all_data()
num_classes = len(label_map)

print(f"[{time.strftime('%H:%M:%S')}] –ó–∞–≥—Ä—É–∂–µ–Ω–æ {len(data)} –ø—Ä–∏–º–µ—Ä–æ–≤")
print(f"–ö–ª–∞—Å—Å–æ–≤: {num_classes}")
print(f"\n–†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –ø–æ –∫–ª–∞—Å—Å–∞–º:")
for orig_label, mapped in sorted(label_map.items()):
    count = sum(1 for d in data if d['original_label'] == orig_label)
    print(f"   –ö–ª–∞—Å—Å {orig_label} (mapped: {mapped}): {count} –ø—Ä–∏–º–µ—Ä–æ–≤")

print("\n" + "=" * 60)
print("–û–ë–£–ß–ï–ù–ò–ï CatBoost_Native_Text –ù–ê –í–°–Å–ú –î–ê–¢–ê–°–ï–¢–ï")
print("=" * 60)

texts = [d['text_lemma'] for d in data]
labels = [d['label'] for d in data]

train_pool = Pool(
    data=[[t] for t in texts],
    label=labels,
    text_features=[0]
)

print(f"\n[{time.strftime('%H:%M:%S')}] –û–±—É—á–µ–Ω–∏–µ –º–æ–¥–µ–ª–∏...")
start_time = time.time()

model = CatBoostClassifier(
    random_seed=SEED,
    iterations=500,
    verbose=100,
    allow_writing_files=False,
    task_type="GPU", 
    text_features=[0]
)

model.fit(train_pool)
train_time = time.time() - start_time

print(f"\n[{time.strftime('%H:%M:%S')}] –û–±—É—á–µ–Ω–∏–µ –∑–∞–≤–µ—Ä—à–µ–Ω–æ –∑–∞ {train_time:.1f}s")

print(f"\n[{time.strftime('%H:%M:%S')}] –°–æ—Ö—Ä–∞–Ω–µ–Ω–∏–µ –º–æ–¥–µ–ª–∏...")

model.save_model(os.path.join(MODEL_DIR, "catboost_model.cbm"))

with open(os.path.join(MODEL_DIR, "label_map.json"), "w", encoding='utf-8') as f:
    json.dump({
        "label_map": {str(k): v for k, v in label_map.items()},
        "reverse_label_map": {str(k): v for k, v in reverse_label_map.items()}
    }, f, ensure_ascii=False, indent=2)

print(f"   –ú–æ–¥–µ–ª—å —Å–æ—Ö—Ä–∞–Ω–µ–Ω–∞ –≤: {MODEL_DIR}/catboost_model.cbm")

print("\n" + "=" * 60)
print("–¢–ï–°–¢–ò–†–û–í–ê–ù–ò–ï –ù–ê –í–°–Å–ú –î–ê–¢–ê–°–ï–¢–ï")
print("=" * 60)

predictions = model.predict(train_pool)
predictions = np.array(predictions).flatten().astype(int)
probabilities = model.predict_proba(train_pool)

accuracy = accuracy_score(labels, predictions)
f1_macro = f1_score(labels, predictions, average='macro')
f1_weighted = f1_score(labels, predictions, average='weighted')

print(f"\n–û–ë–©–ò–ï –ú–ï–¢–†–ò–ö–ò:")
print(f"   Accuracy:    {accuracy:.4f} ({accuracy*100:.2f}%)")
print(f"   F1 Macro:    {f1_macro:.4f}")
print(f"   F1 Weighted: {f1_weighted:.4f}")

print("\n" + "=" * 60)
print("–û–¢–ß–Å–¢ –ü–û –ö–õ–ê–°–°–ê–ú")
print("=" * 60)

class_names = [f"Class_{reverse_label_map[i]}" for i in range(num_classes)]

report = classification_report(
    labels, predictions,
    target_names=class_names,
    digits=4
)
print(report)

print("\n" + "=" * 60)
print("–ê–ù–ê–õ–ò–ó –û–®–ò–ë–û–ö")
print("=" * 60)

cm = confusion_matrix(labels, predictions)

plt.figure(figsize=(12, 10))
sns.heatmap(
    cm, 
    annot=True, 
    fmt='d', 
    cmap='Blues',
    xticklabels=class_names,
    yticklabels=class_names
)
plt.title('–ú–∞—Ç—Ä–∏—Ü–∞ –æ—à–∏–±–æ–∫ CatBoost_Native_Text', fontsize=14)
plt.xlabel('–ü—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–Ω—ã–π –∫–ª–∞—Å—Å', fontsize=12)
plt.ylabel('–ò—Å—Ç–∏–Ω–Ω—ã–π –∫–ª–∞—Å—Å', fontsize=12)
plt.tight_layout()
plt.savefig(os.path.join(MODEL_DIR, "confusion_matrix.png"), dpi=150)

plt.figure(figsize=(12, 10))
cm_normalized = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis] * 100
sns.heatmap(
    cm_normalized, 
    annot=True, 
    fmt='.1f', 
    cmap='Reds',
    xticklabels=class_names,
    yticklabels=class_names
)
plt.title('–ú–∞—Ç—Ä–∏—Ü–∞ –æ—à–∏–±–æ–∫ (% –æ—Ç –∏—Å—Ç–∏–Ω–Ω–æ–≥–æ –∫–ª–∞—Å—Å–∞)', fontsize=14)
plt.xlabel('–ü—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–Ω—ã–π –∫–ª–∞—Å—Å', fontsize=12)
plt.ylabel('–ò—Å—Ç–∏–Ω–Ω—ã–π –∫–ª–∞—Å—Å', fontsize=12)
plt.tight_layout()
plt.savefig(os.path.join(MODEL_DIR, "confusion_matrix_normalized.png"), dpi=150)

errors = []
for i, (true_label, pred_label) in enumerate(zip(labels, predictions)):
    if true_label != pred_label:
        errors.append({
            "index": i,
            "text_raw": data[i]['text_raw'],
            "text_lemma": data[i]['text_lemma'],
            "true_class": reverse_label_map[true_label],
            "pred_class": reverse_label_map[pred_label],
            "true_prob": probabilities[i][true_label],
            "pred_prob": probabilities[i][pred_label]
        })

error_pairs = {}
for e in errors:
    pair = (e['true_class'], e['pred_class'])
    if pair not in error_pairs:
        error_pairs[pair] = []
    error_pairs[pair].append(e)

sorted_pairs = sorted(error_pairs.items(), key=lambda x: len(x[1]), reverse=True)

errors_df = pd.DataFrame(errors)
if len(errors) > 0:
    errors_df.to_csv(os.path.join(MODEL_DIR, "errors_detail.csv"), index=False, encoding='utf-8')

pairs_summary = []
for (true_cls, pred_cls), errs in sorted_pairs:
    pairs_summary.append({
        "True_Class": true_cls,
        "Predicted_Class": pred_cls,
        "Error_Count": len(errs),
        "Avg_True_Prob": np.mean([e['true_prob'] for e in errs]),
        "Avg_Pred_Prob": np.mean([e['pred_prob'] for e in errs])
    })

pairs_df = pd.DataFrame(pairs_summary)
if len(pairs_summary) > 0:
    pairs_df.to_csv(os.path.join(MODEL_DIR, "error_pairs_summary.csv"), index=False)

stats = []
for i in range(num_classes):
    orig_label = reverse_label_map[i]
    total = sum(1 for l in labels if l == i)
    correct = cm[i, i]
    wrong = total - correct
    
    confused_with = []
    for j in range(num_classes):
        if i != j and cm[i, j] > 0:
            confused_with.append((reverse_label_map[j], cm[i, j]))
    confused_with.sort(key=lambda x: x[1], reverse=True)
    
    stats.append({
        "–ö–ª–∞—Å—Å": orig_label,
        "–í—Å–µ–≥–æ": total,
        "–í–µ—Ä–Ω–æ": correct,
        "–û—à–∏–±–æ–∫": wrong,
        "Accuracy": f"{correct/total*100:.1f}%",
        "–ü—É—Ç–∞—é—Ç —Å": ", ".join([f"{c[0]}({c[1]})" for c in confused_with[:3]]) if confused_with else "-"
    })

stats_df = pd.DataFrame(stats)
stats_df.to_csv(os.path.join(MODEL_DIR, "class_statistics.csv"), index=False, encoding='utf-8')

print(f"\n–ò–¢–û–ì–û:")
print(f"   Accuracy: {accuracy*100:.2f}%")
print(f"   –û—à–∏–±–æ–∫: {len(errors)} –∏–∑ {len(data)}")

if errors:
    print("\n" + "=" * 60)
    print("–û–®–ò–ë–û–ß–ù–´–ï –¢–ï–ö–°–¢–´:")
    print("=" * 60)
    for e in errors:
        print(f"\n–ò—Å—Ç–∏–Ω–Ω—ã–π –∫–ª–∞—Å—Å: {e['true_class']} -> –ü—Ä–µ–¥—Å–∫–∞–∑–∞–Ω: {e['pred_class']}")
        print(f"–¢–µ–∫—Å—Ç: {e['text_raw']}")
```
